# Conversão

```python
import xml.etree.ElementTree as ET
import pandas as pd
import os
```

```python
def xml_to_parquet(xml_path, output_path, tag="row"):
    context = ET.iterparse(xml_path, events=("end",))
    _, root = next(context)

    rows = []

    for event, elem in context:
        if elem.tag == tag:
            row = {child.tag: child.text for child in elem}
            rows.append(row)

            root.clear()

    df = pd.DataFrame(rows)

    print("Shape antes de salvar:", df.shape)  # DEBUG

    df.to_parquet(output_path, index=False)
```

- 2018

```python
xml_path = "/home/tatiane/repositorios/brazil-health-data-project/data_pipeline/data/raw/sim/ano_2018_xml/"
output_path = "/home/tatiane/repositorios/brazil-health-data-project/data_pipeline/data/raw/sim/conversion/ano_2018/"
```

```python
xml_folder = "/home/tatiane/repositorios/brazil-health-data-project/data_pipeline/data/raw/sim/ano_2018_xml/"
output_folder = "/home/tatiane/repositorios/brazil-health-data-project/data_pipeline/data/raw/sim/conversion/ano_2018/"

os.makedirs(output_folder, exist_ok=True)

xml_files = sorted([f for f in os.listdir(xml_folder) if f.endswith(".xml")])

for i, xml_file in enumerate(xml_files):
    
    xml_path = os.path.join(xml_folder, xml_file)
    output_path = os.path.join(output_folder, f"part_{i}.parquet")

    print(f"Convertendo: {xml_file}")

    xml_to_parquet(xml_path, output_path)
```
