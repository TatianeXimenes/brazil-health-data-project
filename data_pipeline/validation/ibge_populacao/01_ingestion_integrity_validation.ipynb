# Ingestion and Integrity Validation

- Format and Schema Verification: Confirm that the data is in its original format (JSON, CSV, Parquet) and that the column names and data types match the initial expectation.
- Row Count: Compare the number of rows received with the number of rows sent by the source to ensure that there was no data loss during transport.
- Corruption Check: Check if the files are not corrupted or incomplete (files with a size of 0 or less).
- Identificação de Nulos: Verificar se campos obrigatórios (primary keys, IDs) estão vindo como null.

```python
import pandas as pd
```

```python
# Leitura

PATH = '/home/tatiane/repositorios/brazil-health-data-project/data_pipeline'
df = pd.read_csv(f"{PATH}/data/raw/ibge_populacao/tabela6579.csv", sep=";", skiprows=3)
```

```python
print('Quantidade de registros e colunas:' , df.shape)
```

```python
resumo = pd.DataFrame({
    'tipo': df.dtypes,
    'nulos': df.isnull().sum(),
    'perc_nulos': (df.isnull().sum() / len(df)) * 100,
    'unicos': df.nunique()
})

resumo
```
