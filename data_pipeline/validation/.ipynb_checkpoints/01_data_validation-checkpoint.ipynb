# Conversão

```python
import pandas as pd
import glob
import xml.etree.ElementTree as ET
import os
import gc
```

- 2018

```python
def parse_xml_iter(file):
    data = []
    
    for event, elem in ET.iterparse(file, events=("end",)):
        if elem.tag == "ROW":  # ajuste conforme estrutura do SIM
            record = {child.tag.lower(): child.text for child in elem}
            data.append(record)
            elem.clear()  # libera memória
    
    return pd.DataFrame(data)

files = glob.glob("/home/tatiane/repositorios/brazil-health-data-project/data_pipeline/data/raw/sim/ano_2018_xml/*.xml")
```

```python
output_dir = "/home/tatiane/repositorios/brazil-health-data-project/data_pipeline/data/processed/sim/ano_2018/"
os.makedirs(output_dir, exist_ok=True)
```

```python
for i, file in enumerate(files):
    print(f"Processando {file}...")

    df = parse_xml_iter(file)
    df["ano"] = 2018

    df.to_parquet(f"{output_dir}/part_{i}.parquet", index=False)

    # 🔥 Libera memória manualmente
    del df
    gc.collect()
```

- Todos os anos

```python
import os
import glob

raw_base = "/home/tatiane/repositorios/brazil-health-data-project/data_pipeline/data/raw/sim/"

anos = []

for pasta in os.listdir(raw_base):
    if not pasta.startswith("ano_"):
        continue
    
    # extrair o ano
    try:
        ano = int(pasta.split("_")[1])
    except:
        continue
    
    # começar a partir de 2019
    if ano < 2019:
        continue
    
    path = os.path.join(raw_base, pasta)
    xml_files = glob.glob(f"{path}/*.xml")
    
    # ignorar pastas vazias
    if len(xml_files) == 0:
        continue
    
    anos.append((ano, pasta))

# ordenar por ano
anos = sorted(anos)

print("Anos encontrados:", anos)
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
import gc

for ano, pasta in anos:
    print(f"\n Processando {ano}...")

    input_path = os.path.join(raw_base, pasta)
    output_path = f"/home/tatiane/repositorios/brazil-health-data-project/data_pipeline/data/processed/sim/ano_{ano}/"
    os.makedirs(output_path, exist_ok=True)

    files = glob.glob(f"{input_path}/*.xml")

    for i, file in enumerate(files):
        print(f"  → Arquivo {i+1}/{len(files)}")

        df = parse_xml_iter(file)
        df["ano"] = ano

        df.to_parquet(f"{output_path}/part_{i}.parquet", index=False)

        del df
        gc.collect()
```
