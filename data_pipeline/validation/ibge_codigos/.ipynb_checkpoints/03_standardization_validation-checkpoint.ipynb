# Validation

```python
import pandas as pd
```

```python
# Paths
PATH = '/home/tatiane/repositorios/brazil-health-data-project/data_pipeline'
conversion_path = f"{PATH}/data/processed/ibge_codigos/01_conversion/dtb_brasil_2024_municipios.parquet"
standardization_path = f"{PATH}/data/processed/ibge_codigos/02_standardization/dtb_br_2024_municipios.parquet"
```

```python
# Reading
conversion = pd.read_parquet(conversion_path)
standardization = pd.read_parquet(standardization_path)
```

- Validation

```python
resumo_conversion = pd.DataFrame({
    'tipo': conversion.dtypes,
    'nulos': conversion.isnull().sum(),
    'perc_nulos': (conversion.isnull().sum() / len(conversion)) * 100,
    'unicos': conversion.nunique()
})

resumo_conversion
```

```python
resumo_standardization = pd.DataFrame({
    'tipo': standardization.dtypes,
    'nulos': standardization.isnull().sum(),
    'perc_nulos': (standardization.isnull().sum() / len(standardization)) * 100,
    'unicos': standardization.nunique()
})

resumo_standardization
```
