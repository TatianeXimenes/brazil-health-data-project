# Validation

```python
import pandas as pd
```

```python
# Paths
PATH = '/home/tatiane/repositorios/brazil-health-data-project/data_pipeline'
xls_path = f"{PATH}/data/raw/ibge_codigos/RELATORIO_DTB_BRASIL_2024_MUNICIPIOS.xls"
parquet_path = f"{PATH}/data/processed/ibge_codigos/01_conversion/dtb_brasil_2024_municipios.parquet"
```

```python
# Reading
df_xls = pd.read_excel(xls_path)
df_parquet = pd.read_parquet(parquet_path)
```

- Validation

```python
comparacao = pd.concat([
    df_xls.dtypes.rename('tipo_xls'),
    df_parquet.dtypes.rename('tipo_parquet'),
    df_xls.isnull().sum().rename('nulos_xls'),
    df_parquet.isnull().sum().rename('nulos_parquet'),
    df_xls.nunique().rename('unicos_xls'),
    df_parquet.nunique().rename('unicos_parquet')
], axis=1)

comparacao['tipo_ok'] = comparacao['tipo_xls'] == comparacao['tipo_parquet']
comparacao['nulos_ok'] = comparacao['nulos_xls'] == comparacao['nulos_parquet']
comparacao['unicos_ok'] = comparacao['unicos_xls'] == comparacao['unicos_parquet']
```

```python
comparacao.query("tipo_ok == False or nulos_ok == False or unicos_ok == False")
```

- Visão completa

```python
resumo_xls = pd.DataFrame({
    'tipo': df_xls.dtypes,
    'nulos': df_xls.isnull().sum(),
    'perc_nulos': (df_xls.isnull().sum() / len(df_xls)) * 100,
    'unicos': df_xls.nunique()
})

resumo_xls
```

```python
resumo_parquet = pd.DataFrame({
    'tipo': df_parquet.dtypes,
    'nulos': df_parquet.isnull().sum(),
    'perc_nulos': (df_parquet.isnull().sum() / len(df_parquet)) * 100,
    'unicos': df_parquet.nunique()
})

resumo_parquet
```
