# Standardization

```python
import pandas as pd
import re
```

## Leitura do arquivo

```python
df = pd.read_csv(
    "/home/tatiane/repositorios/brazil-health-data-project/data_pipeline/data/raw/ibge_populacao/tabela6579.csv",
    sep=";",
    skiprows=3
)
```

```python
print(len(df))
df.head()
```

```python
# Tranformando wide - long
df_long = df.melt(
    id_vars=["Cód.", "Unidade da Federação"],
    var_name="ano",
    value_name="populacao"
)
```

```python
print(len(df_long))
df_long.head()
```

```python
df_long.columns = df_long.columns.str.strip()
```

### Writing the dataset

```python
df_long.to_parquet('/home/tatiane/repositorios/brazil-health-data-project/data_pipeline/data/analytics/ibge_populacao/ibge_populacao.parquet', index=False)
```
